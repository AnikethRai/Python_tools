### Array types and Conversions between types


```python
import numpy as np

#To create array in numpy
integers = np.array([1,2,3,4,5,6])

#To check type of array
print(f'integers dtype is {integers.dtype}')

#To create array with short integers in size
integers_short = np.array(integers, dtype=np.int8)

#Checking between integers and integers_short
print(f'intergers bytes : {integers.nbytes} ||| integers_short : {integers_short.nbytes}')
```

### Intrinsic Numpy Array creation

### Adding , removing and sorting 

```python
#Inserting or adding new element at a particular position.
test_arr = np.array([1,2,3,5])
#insert (arr, loc, value)
new_arr = np.insert(test_arr , 3 , 4)
print(new_arr)


#Appending value to array (arr , value)
s_arr = np.append(new_arr , 6)
print(s_arr)

#Deleting an element (arr , loc)
d_arr = np.delete(s_arr , 0)
print(d_arr)

#Sorting Arrays
test2_arr = np.array([3,4,1,2,5,6])
print(np.sort(test2_arr))
# In strings
test3_arr = np.array(['strings' , 'int' , 'alpha' , 'aes' , 'blue' , 'cat'])
print(np.sort(test3_arr))
```

### Copies and views


```python
#Copies create new address with same elements of a copied array
arra = np.array([1,2,3,4,5])
print('Before using copying function')
arra_2 = arra
print(f'arra address : {id(arra)} \n arra_2 address : {id(arra_2)}')
print('After using copying function')
copied_arra = arra.copy()
print(f'arra address : {id(arra)} \n copied_arra address : {id(copied_arra)}')
print(copied_arra , arra)

#After chaning value in arra the original array
arra[0] = 1000
print('after changing value in original array')
print(copied_arra , arra)



#Creating Views -> in this if change either original or view data both data changes.

view_arra = arra.view()
print(f'arra address : {id(arra)} \n view_arra address : {id(view_arra)}')
print(view_arra , arra)

#After chaning value in arra the original array
arra[0] = 1000
print('after changing value in original array')
print(view_arra , arra)
```

### Joining and Splitting arrays

concatenate , stack , hstack, and vstack
split , array_split , hsplit , vsplit

```python
#Concatenate

arra1 = np.arange(1,11)
arra2 = np.arange(11,21)
con_ara = np.concatenate((arra1,arra2))
print(con_ara)

#stack
stk_ara = np.stack((arra1 , arra2))
print(stk_ara)

#hstack
hstk_ara = np.hstack((arra1 , arra2))
print(hstk_ara)

#vstack
vstk_ara = np.vstack((arra1 , arra2))
print(vstk_ara)


##############################
#split
sp_arr = np.array_split(arra1  , 2)
print(sp_arr)

```

#### Aggregate functions
        

```python
ar1 = np.arange(10,110,10)
ar2 = np.arange(10,100,10).reshape(3,3)
ar3= np.arange(10,110,10).reshape(2,5)

#sum
print(np.sum(ar1))
print(np.sum(ar2,axis = 0))
print(np.sum(ar3))

#prod
print(np.prod(ar1))
print(np.prod(ar2,axis = 0))
print(np.prod(ar3))

#average
print(np.average(ar1))
print(np.average(ar2,axis = 0))
print(np.average(ar3))

#min , max , standard deviation
print(np.min(ar1))
print(np.max(ar1))
print(np.std(ar1))

```

# Application - Linear Regression

# House market
- Input: Price data from 2012 - 2021
- Output: Avarage house market 2022?
- Asume Relationship - squared
- y=ax<sup>2</sup>+bx+c

```python
year = np.arange(1,11)
price = np.array([129000, 133000, 138000, 144000, 142000, 141000, 150000, 135000, 134000, 137000]) 
year 
```

```python
a, b, c = np.polyfit(year, price, 2) 
print ("a:",a)
print ("b:",b)
print ("c:",c)
```

```python
print("Estimated price for 2022:",a*11**2 + b*11 + c )
plt.plot(year,price, color = 'blue')
plt.scatter(year,price, color = 'blue')
plt.scatter(11, a*11**2 + b*11 + c ,color='red')
plt.title('Linear regression')
plt.xlabel('year')
plt.ylabel('average house price')
```
